Creating clusters from Sklearn to understand customers' behaviours

```python
X1 = df_csa[['MntTotal','Income']]
X1
```

```python

from sklearn.cluster import KMeans
inertias = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, random_state=42, init='k-means++')
    kmeans.fit(X1)
    inertias.append(kmeans.inertia_)
plt.figure(figsize=(10,6))
plt.plot(range(1,11), inertias, color='orange')
plt.xlabel('K Values')
plt.ylabel('inertias')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
Cluster = kmeans.fit_predict(X1)
#print (labels)
center = kmeans.cluster_centers_
center
```

```python
X1['Cluster'] = kmeans.predict(X1)
X1
```

```python
plt.scatter(X1['MntTotal'],X1['Income'], c = kmeans.labels_)
plt.grid()
plt.xlabel('MntTotal')
plt.ylabel('Income')
plt.scatter(center[:,0],center[:,1], color = 'black')
plt.title('Cluster of Customers by Total Amount and Income')
```

```python
joined_df_csa = df_csa.join(X1['Cluster'], how='outer')
joined_df_csa
```

Creating New column for childrenhome and joining 

```python
def childrenhome(row):
    if row['Kidhome'] == 1:
        return 1
    elif row['Teenhome'] ==1:
        return 1
    else:
        return 0
joined_df_csa['Childrenhome'] = joined_df_csa.apply(childrenhome, axis = 1)
```
